```python

for make in makes:
    for province in provinces:


        firmwebsite ="https://www.dealerrater.ca/directory/"+province+"/"+make+"/page"

        for x in range (1,10):

            website =firmwebsite+str(x)

            driver.get(website)

            xpath = "//*[@id='dealerName']//*[contains(@class,'teal boldest font-20 uppercase dealer-name-link')]"
            names=driver.find_elements_by_xpath(xpath)

            xpath = "//*[@id='dealerName']//*[contains(@class,'col-xs-12 pad-none text-left no-wrap hidden-xs')]"
            locations=driver.find_elements_by_xpath(xpath)

            xpath = "//*[@id='dealerName']//*[@href]"
            links=driver.find_elements_by_xpath(xpath)

            for i in range(0,len(names)):
                name = names[i].text
                location = locations[i].text
                link = links[i].get_attribute('href')
                onefirm_data=pd.DataFrame([[name,location,province,make,link]],columns=dfcolumns1)
                data1=data1.append(onefirm_data)

            if len(names)==0:
                break

```
