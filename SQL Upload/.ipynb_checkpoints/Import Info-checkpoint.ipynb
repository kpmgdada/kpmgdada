```python
import pandas as pd
import os
import xlrd 
from os.path import splitext, basename
from re import search
import numpy 
dir_name = 'C:\Database'
outputfilename='FinalData.csv'



def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

def threshold(a, x):
     return sorted(item for item in a if item == x)
    
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
UniqueData=newData.groupby(['ServiceArea', 'Asset']).size().reset_index(name='Freq')
LOSNames=['LOS1','LOS2','LOS3','LOS4','LOS5','LOS6','LOS7','LOS8','LOS9','LOS10','LOS11','LOS12']
ProbName=['Prob-'+s for s in LOSNames]
ExtentName=['Extent-'+s for s in LOSNames]
CLOSName=['CLOS-'+s for s in LOSNames]

FinalData = pd.DataFrame(columns=columnsData)


for i in range(len(UniqueData)):
    AssetName=UniqueData['Asset'].iloc[i]
    SGName=UniqueData['ServiceArea'].iloc[i]
    dfTemp=newData.loc[newData['Asset']==AssetName]
    dfTemp=dfTemp.loc[dfTemp['ServiceArea']==SGName]

    WeightCur=Weight.loc[(Weight['Asset'] == AssetName) & (Weight['ServiceArea'] == SGName)]

    ProbList=(WeightCur[ProbName].values.tolist()[0])
    ProbList=[str(i) for i in ProbList]
    ProbList=['Prob'+s for s in ProbList]
    ProbData=dfTemp[ProbList]


    ExtentList=(WeightCur[ExtentName].values.tolist()[0])
    ExtentList=[str(i) for i in ExtentList]
    ExtentList=['Extent'+s for s in ExtentList]
    ExtentData=dfTemp[ExtentList]

    LOSData=dfTemp[LOSNames]

    CLOSList=(WeightCur[CLOSName].values.tolist()[0])
    CLOSList=[str(i) for i in CLOSList]
    CLOSList=['CLOS'+s for s in CLOSList]

    calData=ExtentData.values*ProbData.values*LOSData.values*WeightCur[LOSNames].values
    calData = pd.DataFrame(calData, columns = LOSNames)

    indicesCLOS1 = [i for i, x in enumerate(CLOSList) if x == 'CLOS1']
    indicesCLOS2 = [i for i, x in enumerate(CLOSList) if x == 'CLOS2']
    indicesCLOS3 = [i for i, x in enumerate(CLOSList) if x == 'CLOS3']
    indicesCLOS4 = [i for i, x in enumerate(CLOSList) if x == 'CLOS4']

    sumCLOS1=sum(calData.T.iloc[indicesCLOS1])
    sumCLOS2=sum(calData.T.iloc[indicesCLOS2])
    sumCLOS3=sum(calData.T.iloc[indicesCLOS3])
    sumCLOS4=sum(calData.T.iloc[indicesCLOS4])
    sumTotCLOS=sumCLOS1+sumCLOS2+sumCLOS3+sumCLOS4

    dfTemp['CLOS1']=sumCLOS1.as_matrix() 
    dfTemp=dfTemp.assign(CLOS2=sumCLOS2.as_matrix() )
    dfTemp=dfTemp.assign(CLOS3=sumCLOS3.as_matrix() )
    dfTemp['CLOS4']=sumCLOS4.as_matrix() 
    dfTemp=dfTemp.assign(TotCLOS=sumTotCLOS.as_matrix() )
    FinalData=FinalData.append(dfTemp)
```

```python

FinalData.to_csv (os.path.join(dir_name, outputfilename), index = None, header=True)
```
